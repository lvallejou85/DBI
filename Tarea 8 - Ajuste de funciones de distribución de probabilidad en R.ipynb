## Tarea 8 - Ajuste de funciones de distribución de probabilidad en R

### Paso 1: Importar y cargar las librerías requeridas

```r
library(MASS)
library(vcd)
library(fitdistrplus)
options(warn=-1) # suprime los warnings
```

### Paso 2: Lectura de datos

```r
#Lea los datos del archivo "Datos" y asígnelos a un objeto cuyo nombre sea "datos"

#Sugerencia: Aunque el archivo está en formato .txt, es posible usar la función read.csv.
#Escriba el argumento sep="\n" para que R reconozca que cada entrada está en un renglón distinto

datos <- read.csv("Datos distribuciones.txt", sep="\n", col.names = c("value"))
head(datos)
```

```r
#Convierta el data frame en un vector

#Sugerencia: Use el comando as.numeric

datos <- as.numeric(datos$value)
datos
```

```r
#Una vez tenga el vector de datos, haga un histograma para examinar alguna posible función de distribución
#de probabilidad. Asigne este histograma a un objeto h
h <- hist(datos, freq = FALSE, main="Distribución", ylab = "Frecuencia")
d <- density(datos)
lines(d, col = 'red', lwd = 1)
summary(datos)
```

```r
#Use la función fitdistr(conjunto de datos, distribucion a ajustar) para explorar algunas posibles funciones de distribución 
#de probabilidad. Ensaye con la gamma, lognormal y normal

#Haga un gráfico de los datos vs cada distribución ajustada
```

```r
gamma <- fitdist(datos, "gamma")
gamma
```

```r
denscomp(gamma, fitcol="blue", datacol="white", main="Real vs Gamma")
lines(d, col = 'red')
legend("topright", c( "Real","Ajustado","Densidad"), fill=c("white", "blue", "red"))
```

```r
lognormal <- fitdist(datos, "lnorm")
lognormal
```

```r
denscomp(lognormal, fitcol="blue", datacol="white", main="Real vs Lognormal")
lines(d, col = 'red')
legend("topright", c( "Real","Ajustado","Densidad"), fill=c("white", "blue", "red"))
```

```r
normal <- fitdist(datos, "norm")
normal
```

```r
denscomp(normal, fitcol="blue", datacol="white", main="Real vs Normal")
lines(d, col = 'red')
legend("topright", c( "Real","Ajustado","Densidad"), fill=c("white", "blue", "red"))
```
